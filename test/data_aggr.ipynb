```python
import pandas as pd
import numpy as np


raw_data_path = "/home/tongyu/project/hperf/tmp/20221207_test006/perf_result"
raw_dataframe = pd.read_csv(raw_data_path,
                            header=None,
                            names=["timestamp", "unit", "value", "event"],
                            usecols=[0, 1, 2, 4])
raw_dataframe
```

```python
event_per_cpu = raw_dataframe.groupby(["unit", "event"]).agg(
    value_per_cpu=("value", np.sum)
)
event_per_cpu
```

```python
event_all_cpu_total = event_per_cpu.groupby(["event"]).agg(
    value_all_cpu=("value_per_cpu", np.sum)
)
event_all_cpu_total
```
