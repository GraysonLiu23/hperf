```python
perf_raw_data = pd.read_csv(os.path.join(test_dir, "perf_result"),
                                    sep="\t",
                                    header=None, 
                                    names=["timestamp", "unit", "value", "metric"], 
                                    usecols=[0, 1, 2, 4])

perf_raw_data
```

```python
def cpu2socket(x):
    cpu_id = int(x["unit"][3:])
    return "SOCKET" + str(cpu_topo.loc[cpu_topo.unit == cpu_id]["socket"].values[0])

perf_raw_data.loc[perf_raw_data.metric == "uncore_cha_0/event=0x34,umask=0x1fe001/", ["unit"]] = perf_raw_data.loc[perf_raw_data.metric == "uncore_cha_0/event=0x34,umask=0x1fe001/", ["unit"]].apply(cpu2socket, axis=1)
```
